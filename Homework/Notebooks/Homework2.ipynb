```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im=cv.imread('D:\Semester 4\Image processing\Homework\Homework2\spider.png')
assert im is not None
cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',im)
cv.waitKey(0)
gamma = [0.2,0.8,1,1.2,2]
for i in gamma:
    t = np.array([(p/255)**i*255 for p in range (0,256)]).astype(np.uint8)
    g = cv.LUT(im,t)
    rgb_im = cv.cvtColor(g, cv.COLOR_BGR2RGB)

    fig,ax = plt.subplots()
    ax.plot(t)
    

    cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
    cv.imshow('Image',g)
    cv.waitKey(0)
    cv.destroyAllWindows()    

    fig = plt.subplots()
    plt.imshow(rgb_im)
    plt.show()
```

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

im=cv.imread('D:\Semester 4\Image processing\Homework\Homework2\shells.tif',cv.IMREAD_GRAYSCALE)
assert im is not None

hist_im = cv.calcHist([im], [0], None, [256], [0,256])
g = cv.equalizeHist(im)
hist_g = cv.calcHist([g], [0], None, [256], [0,256])

fig, ax = plt.subplots(2,1,figsize= (8,8))
ax[0].plot(hist_im)
ax[1].plot(hist_g)

cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',g)
cv.waitKey(0)
cv.destroyAllWindows() 

rgb_im = cv.cvtColor(g, cv.COLOR_BGR2RGB) 
fig = plt.subplots()
plt.imshow(rgb_im)
plt.show()
```

```python
im=cv.imread('D:\Semester 4\Image processing\Homework\Homework2\zion_pass.jpg')
assert im is not None

hsv_im = cv.cvtColor(im, cv.COLOR_BGR2HSV)

(h, s, v) = cv.split(hsv_im)
s = s*2
s = np.clip(s,0,255)
hsv2_im = cv.merge([h,s,v])

bgr_im1 = cv.cvtColor(hsv2_im, cv.COLOR_HSV2BGR)
cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',bgr_im1)
cv.waitKey(0)

(h, s, v) = cv.split(hsv_im)
h = h*2
h = np.clip(h,0,255)
hsv3_im = cv.merge([h,s,v]) 

bgr_im2 = cv.cvtColor(hsv3_im, cv.COLOR_HSV2BGR)
cv.namedWindow('Image',cv.WINDOW_AUTOSIZE)
cv.imshow('Image',bgr_im2)
cv.waitKey(0)
cv.destroyAllWindows() 

rgb_im1 = cv.cvtColor(bgr_im1, cv.COLOR_BGR2RGB) 
rgb_im2 = cv.cvtColor(bgr_im2, cv.COLOR_BGR2RGB) 
fig = plt.subplots()
plt.imshow(rgb_im1)
plt.show()

plt.imshow(rgb_im2)
plt.show()
```
